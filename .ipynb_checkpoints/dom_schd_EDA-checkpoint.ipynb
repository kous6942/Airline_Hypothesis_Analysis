```python
#calculating Coefficient of Variation to understand degree of variablity amongst passengers, ASK and Load Factor
#passengers:
'''
CV Value	Meaning
< 5%	Very stable
5–15%	Mild volatility
> 15%	Highly volatile
> 30%	Extremely volatile
'''

CV_Pass = (192015.1/2152244)*100
print('CV for passengers =: ', CV_Pass)
print('CV bw 5-15% indicates mild volatility')
```

```python
#available seat km:
CV_Pass = (147397.5/2836663)*100
print('CV for available seat KM =: ', CV_Pass)
print('CV being 5% indicates stability in ASK')
```

```python
#PAX Load Factor:
CV_Pass = (2.896310/81.488889)*100
print('CV for PAX =: ', CV_Pass)
print('CV being <5% indicates strong stability in PAX')
```
